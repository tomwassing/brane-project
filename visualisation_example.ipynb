We need both the matplotlib and sklearn packages

```branescript
import matplotlib;
import sklearn;
```

We load the iris dataset and the corresponding colormap.

```branescript
let data := load_iris("data");
let target := load_iris("target");
```

We plot the data extracted from the iris dataset.

```branescript
print(plot_file(data, "/data/standard_dataset.png", true, "Sepal length", "Sepal width", "PCA on iris Dataset", target, "Set1", "k"));
```
